### Type Effectiveness Columns (Against Fire, Against Water, etc.)

These columns represent the damage multiplier a Pokémon receives when attacked by a specific type. They indicate the Pokémon's weaknesses, resistances, and immunities.

- 0 → Immune (No damage)
- 0.25 → Very Resistant (¼ damage)
- 0.5 → Resistant (½ damage)
- 1 → Normal damage
- 2 → Weak (2× damage)
- 4 → Very Weak (4× damage)

Against Fire = 2 → Fire-type attacks make it lose 2× more HP (weakness).<br>
Against Water = 0.5 → Water-type attacks make it lose half as much HP (resistant).<br>
Against Ground = 0 → Ground-type attacks cause no HP loss (immune).<br>

Final Evolution → Last stage of evolution.<br>
Legendary → Rare and powerful Pokémon.<br>
Mega Evolution → Temporary powerful form during battle.<br>
Alolan Form → Alola region version.<br>
Galarian Form → Galar region version.<br>

### 📊 Statistical Metrics

#### 🔹 BST (Base Stat Total)
BST is the sum of a Pokémon's six battle stats (HP, Attack, Defense, Special Attack, Special Defense, and Speed). It represents the Pokémon's overall strength.

#### 🔹 Mean
The mean is the average of a Pokémon's six battle stats. It shows the average strength of an individual stat.

#### 🔹 Standard Deviation (Std)
Standard Deviation measures how balanced a Pokémon's six stats are. A low value means the stats are evenly distributed, while a high value means the Pokémon has some very high and some very low stats.

#### 🔹 Z-Score
The Z-score compares a Pokémon's BST with the average BST of all Pokémon in the dataset. It tells us whether a Pokémon's overall strength is below average, average, or above average.

- Z-Score < 0 → Below average
- Z-Score = 0 → Average
- Z-Score > 0 → Above average

#### Business Question

**Has the average power of Pokémon changed across generations?**


#### Metric Used

**Power** is measured using **BST (Base Stat Total)**, which is the sum of a Pokémon's **HP, Attack, Defense, Special Attack, Special Defense, and Speed**.


#### Findings

- The average BST remained almost unchanged between **Generation 1 (407.64)**, **Generation 2 (407.18)**, and **Generation 3 (407.99)**.
- A noticeable increase occurred in **Generation 4**, where the average BST increased to **451.93**.
- The average BST then declined in **Generation 5** to **433.85**.
- **Generation 6** introduced the strongest Pokémon on average, with the highest average BST of **508.67**.
- After Generation 6, the average BST decreased to **450.64** in **Generation 7** and **443.83** in **Generation 8**.


##### Business Insight

The analysis suggests that Pokémon power did **not** consistently increase across generations. Instead, there were periods of both increases and decreases. **Generation 6** stands out as the generation with the highest average Pokémon strength, while **Generations 1–3** maintained a relatively stable average BST.
